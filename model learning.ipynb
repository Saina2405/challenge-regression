```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('data/dataset_regression_cleaned.csv')
```

```python
# define a function called "outliers" which returns a list of index of outliers
# Use Inter-quartile Range(IQR) to detect ourliers
# IQR = Q3-Q1
# +/- 1.5*IQR

def outliers(df, variable):
    Q1 = df[variable].quantile(0.30)
    Q3 = df[variable].quantile(0.70)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR

    ls = df.index[(df[variable] < lower_bound) | (df[variable] > upper_bound)]
    return ls 
```

```python
# create an empty list to store the output indice from multiple columns
index_list = []
for variable in ['Price']:
    index_list.extend(outliers(df, variable))
```

```python
# define a function called "remove" which returns a cleaned dataframe without outliers
def remove(df, ls):
    ls = sorted(set(ls)) 
    Newdf = df.drop(ls)
    return df
```
